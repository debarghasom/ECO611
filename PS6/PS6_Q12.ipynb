```python
import numpy as np
import pandas as pd
import scipy.stats as st
np.random.seed(100)

def ks_test(sample1, sample2):
    # Sort the samples
    sample1_sorted = np.sort(sample1)
    sample2_sorted = np.sort(sample2)
    
    # Calculate the CDFs of the samples
    cdf1 = np.searchsorted(sample1_sorted, sample1, side='right') / len(sample1_sorted)
    cdf2 = np.searchsorted(sample2_sorted, sample1, side='right') / len(sample2_sorted)
    
    # Calculate the maximum absolute difference between the CDFs
    ks_statistic = np.max(np.abs(cdf1 - cdf2))
    return ks_statistic

data = pd.read_csv("data3.csv",header=None)
#print(data)
x1=data.iloc[:,0]
x2=data.iloc[:,1]
x3=data.iloc[:,2]

mle1=np.mean(x1)/9
mle2=np.mean(x2)/9
mle3=np.mean(x3)/4

smp1=np.random.binomial(9,mle1,10000)
smp2=np.random.binomial(9,mle1,10000)
smp3=np.random.negative_binomial(4,0.573065,10000)


print(ks_test(x1,smp1))
print(ks_test(x2,smp2))
print(ks_test(x3,smp3))

Dn1, p1 = st.kstest(x1,smp1)
Dn2, p2 = st.kstest(x2,smp2)
Dn3, p3 = st.kstest(x3,smp3)
print("scipy",Dn1,p1)
print("scipy",Dn2,p2)
print("scipy",Dn3,p3)
```
