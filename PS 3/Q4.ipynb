```python
import numpy as np
import matplotlib.pyplot as plt

y=[]
x=np.random.randint(1,11,1000)
for i in range(len(x)):
    if x[i] <= 1:
        y.append(1)
    elif x[i] <=3:
        y.append(2)
    elif x[i] <=6:
        y.append(3)
    else:
        y.append(4)
plt.hist(y,bins=10,density=True)

plt.title('Histogram of X')
plt.xlabel('X values')
plt.ylabel('Probability density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fourb():
    x = np.random.rand(1000)
    a = 1 / ((1 ** 0.5) + (2 ** 0.5) + (3 ** 0.5) + (4 ** 0.5))
    b = (2 ** 0.5) / ((1 ** 0.5) + (2 ** 0.5) + (3 ** 0.5) + (4 ** 0.5))
    c = (3 ** 0.5) / ((1 ** 0.5) + (2 ** 0.5) + (3 ** 0.5) + (4 ** 0.5))
    b = b + a
    c = c + b
    for i in range(len(x)):
        if x[i] < a:
            x[i] = 1
        elif a <= x[i] < b:
            x[i] = 2
        elif b <= x[i] < c:
            x[i] = 3
        else:
            x[i] = 4
    plt.hist(x, bins=4, density=True)
    plt.show()
    print(x)

fourb()
```
