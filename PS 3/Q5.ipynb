```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.uniform(-1,1,1000)
c=0
y=[]

for i in x:
    if(i<0):
        c= 1 + i
        y.append(c)
    else:
        c= 1 - i
        y.append(c)
plt.plot(x,y)
plt.show()
plt.scatter(x,y)
plt.show()
plt.hist(y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the inverse transform function
def inverse(u):
    if u <= 0.5:
        return -1 + (2 * u)**0.5
    else:
        return 1 - (2 - 2 * u)**0.5

# Generate 1000 random numbers from a uniform distribution
u_values = np.random.uniform(0, 1, size=1000)

# Apply the inverse transform to get X values
X_values = np.array([])
for u in u_values:
    X_values = np.append(X_values,inverse(u))

# Plot the histogram
plt.hist(X_values, bins=30, density=True, color='blue')

# Plot the true density function fX(x)
#x = np.linspace(-1, 1, 1000)

#def func(i):
 #   if -1 <= i <= 0:
  #      return 1+ i
   # elif 0 <= i <= 1:
    #3    return 1 - i
        
#fX_values = np.array([func(i) for i in x])
#plt.plot(x, fX_values, label='True fX(x)')

plt.title('Histogram and True Density Function')
plt.xlabel('X')
plt.ylabel('Density')
#plt.legend()
plt.show()
```
