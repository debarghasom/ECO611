```python
#1 a)
import sympy as sm
import numpy as np
import matplotlib.pyplot as plt
n=0
x=1
con=np.array([x])

while (np.abs(np.cos(x) - x**3) > 1e-6):
    x = x - (np.cos(x) - x**3)/(-np.sin(x) - 3*x**2)
    con=np.append(con,[x])
    n+=1

y=[i for i in range(n+1)]

plt.plot(con, y, color='red')
plt.show()

print("Approximate root:", x)
print("Iterations of convergence: ",n)
```

```python
#1 b)
import numpy as np
import matplotlib.pyplot as plt
n=0
x=27
con=np.array([x])

while (np.abs(3*(np.cos(x)) - np.exp(x)) > 1e-6):
    x = x - (3*(np.cos(x)) - np.exp(x))/(-3*np.sin(x) - np.exp(x))
    con=np.append(con,[x])
    n+=1

y=[i for i in range(n+1)]

plt.plot(y, con, color='red')
plt.show()

print("Approximate root:", x)
print("Iterations of convergence: ",n)
```

```python
#1 c)
import numpy as np
import matplotlib.pyplot as plt
n=0
x=1
con=np.array([x])

while (np.abs(x**(1/3) - np.exp(-(x**2))) > 1e-6):
    x = x - (x**(1/3) - np.exp(-(x**2)))/((1/3)*x**(-2/3) + 2*x*np.exp(-(x**2)))
    con=np.append(con,[x])
    n+=1

y=[i for i in range(n+1)]

plt.plot(y, con, color='red')
plt.show()

print("Approximate root:", x)
print("Iterations of convergence: ",n)

# x=sm.symbols('x')
# print(f"By sympy: {sm.nsolve(x**(1/3) - sm.exp(-(x**2)),x,0)}")
```
