```python
import numpy as np
import matplotlib.pyplot as plt

#basically wap to generate 1000 y values from 0 to 1 so that F(x) = y, we apply NR method to get back x
x=np.zeros(1000)
def invF(y):
    x=0.5
    while(np.abs(1 - np.exp(-x)*(x+1) - y) > 1e-8):
        x = x - (1 - np.exp(-x)*(x+1) - y)/(x*np.exp(-x))
    return x 

y=np.random.rand(1000)
x=[invF(i) for i in y]

plt.hist(x, bins = 30, density = True, color = 'red')
plt.show

    
```
