```python
import numpy as np
import matplotlib.pyplot as plt
y = np.random.rand(1000)
x=np.zeros(1000)
n = 1
def newton(y,n):
    x0 = 0.5
    if y <= 0.5:
        while ( np.abs(((n+1)/2*n)*(x0 + ((-x0)**(n+1))/(n+1) + n/(n+1)) - y) >= 1e-8):
            x0 = x0 - (((n+1)/2*n)*(x0 + ((-x0)**(n+1))/(n+1) + n/(n+1)) - y)/(((n+1)/2*n)*(1-(-x0)**n))
        return x0
    elif y> 0.5:
        while (np.abs( ((n+1)/2*n)*(x0 - ((x0)**(n+1))/(n+1)) + 0.5 - y) >= 1e-8):
            x0 = x0 - (((n+1)/2*n)*(x0 - ((x0)**(n+1))/(n+1)) + 0.5 - y)/(((n+1)/2*n)*(1-(x0)**n))
        return x0

x=[newton(i,n) for i in y]
plt.hist(x, bins = 30, density = True, color = 'blue')
plt.show
```
