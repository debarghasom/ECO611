```python
#1 c)
import sympy as sm
import numpy as np
import matplotlib.pyplot as plt
n=0
x1 = 3
x2 = 0
x = (x1 + x2) / 2
lhl,rhl=np.array([x1], dtype=float),np.array([x2], dtype=float)

while (np.abs(x**(1/3) - np.exp(-(x**2))) > 1e-6):
    if((x**(1/3) - np.exp(-(x**2))) < 0):
        x2 = x
    elif((x**(1/3) - np.exp(-(x**2))) > 0):
        x1 = x
    x = (x1 + x2) / 2
    n+=1
    lhl = np.append(lhl,[x1])
    rhl = np.append(rhl,[x2])

print("Approximate root:", x)
print("Iterations of convergence: ",n)

y=[i for i in range(n+1)]

plt.plot(y,lhl, color='red', label='Line 1')
plt.plot(y,rhl, color='blue', label='Line 2')
plt.show()

x=sm.symbols('x')
print(f"By sympy: {sm.nsolve(x**(1/3) - sm.exp(-(x**2)),x,0)}")
```
