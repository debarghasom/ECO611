```python
import numpy as np
a,b=(1/3**0.5),(1/2**0.5)
A=np.array([[a,0],[a,b],[a,-b]])
U=np.linalg.svd(A)[0]
m=np.shape(U)[1]

r=np.linalg.matrix_rank(A)
for i in range(r,m):
    U[:,i]=U[:,i]/np.linalg.norm(U[:,i])
    
A=np.concatenate((A,U[:,i:]),axis=1)

print(A)
```
